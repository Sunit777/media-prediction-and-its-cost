```python
for i in df.columns:
    if df[i].dtype=="object":
        df[i]=df[i].astype('category').cat.codes
    
```

```python
df.dtypes
```

```python
x=df.drop('cost',axis=1)
y=df['cost']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.3)
```

```python
x.shape,y.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
model_pred=model.predict(x_test)
model_pred
```

```python
print(f'train accuracy :: {model.score(x_train,y_train)*100:.2f}')
print(f'test accuracy :: {model.score(x_test,y_test)*100:.2f}')
```

```python
from sklearn.tree import DecisionTreeRegressor
model1=DecisionTreeRegressor()
model1.fit(x_train,y_train)
```

```python
model1_pred=model1.predict(x_test)
model1_pred
```

```python
print(f'train accuracy :: {model1.score(x_train,y_train)*100:.2f}')
print(f'test accuracy ::  {model1.score(x_test,y_test)*100:.2f}')
```
